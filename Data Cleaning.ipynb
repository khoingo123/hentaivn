```
# Import necessary libraries
import os
import pandas as pd
import numpy as np

files = os.listdir()
```

```
# Concatenate many single files into one
data = pd.read_csv(files[0])
for file in files[1:]:
    other_data = pd.read_csv(file)
    data = pd.concat([data, other_data], ignore_index=True)
```

## II. Clean data

```
data.head()
```

```
# Drop duplicates
data.drop_duplicates(inplace=True)
```

```
# Preprocess the categories column
def remove_marks(s):
    res = ''
    for c in s:
        if c not in "'[]":
            res += c
    return res

def process_categories(s):
    res = s.split(', ')
    for i in range(len(res)):
        res[i] = remove_marks(res[i])
    return res

data['Categories'] = data['Categories'].apply(process_categories)
```

```
# Preprocess the views column
def process_views(s):
    res = ''
    for c in s:
        if c != '.':
            res += c
    return int(res)

data['Views'] = data['Views'].apply(process_views)
```

```
# Preprocess the Likes and Dislikes column
data['Likes'] = data['Likes'].astype(int)
data['Dislikes'] = data['Dislikes'].astype(int)
```

```
# Preprocess the Date column
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y',errors='coerce')
```
